```python
import tensorflow as tf

import vgg19_trainable as vgg19

import numpy as np
from skimage import io, transform
import glob
import os
import sklearn
import tensorflow as tf
import time
import csv
import cv2 as cv
```

```python
# 将所有的图片resize成100*100
w = 224
h = 224
c = 3
# for x in os.listdir(path):
#     print(x)

# 读取图片
def read_img_train(path):
    w = 100
    h = 100
    imgs = []
    imgs2 = []
    imgs3 = []
    imgs4 = []
    for im in glob.glob(path + '/*.jpg'):
        print('reading the images:%s' % (im))
        img = cv.imread(im,1)
        img = cv.resize(img, (w, h))
        size = img.shape


        iUD = np.zeros(img.shape, np.uint8) 
        iLR = np.zeros(img.shape, np.uint8) 
        iAcross = np.zeros(img.shape, np.uint8) 
        
        for i in range(h):
            for j in range(w):
                iUD[h - 1 - i, j] = img[i, j]
                iLR[i, w - 1 - j] = img[i, j]
                iAcross[h - 1 - i, w - 1 - j] = img[i, j]
        imgs.append(img)
        imgs2.append(iUD)
        imgs3.append(iLR)
        imgs3.append(iAcross)

    csv_file = csv.reader(open('../data/labels.csv', 'r'))

    labels = []
    filenames = []
    for i in csv_file:

        filenames.append(i[0])
        labels.append(i[1])
    filenames = np.array(filenames)
    labels = np.array(labels)

    unique_label = np.unique(labels)

    L = []
    for i in range(len(unique_label)):
        L.append(i)
    L = np.array(L)
    label = []
    for x in os.listdir(path):
        tmp = labels[filenames == x[:-4]]

        tmp2 = L[unique_label == tmp]

        label.append(tmp2[0])

    label = np.array(label)

    return np.asarray(imgs + imgs2 + imgs3 + imgs4, np.float32), np.asarray(np.hstack((label, label, label, label)),np.int32)

def read_img_test(path):

    imgs = []

    for im in glob.glob(path + '/*.jpg'):
        print('reading the images:%s' % (im))
        img = io.imread(im)

        img = transform.resize(img, (w, h), 1, 'constant')


        imgs.append(img)


    csv_file = csv.reader(open('../data/labels.csv', 'r'))

    labels = []
    filenames = []
    for i in csv_file:

        filenames.append(i[0])
        labels.append(i[1])
    filenames = np.array(filenames)
    labels = np.array(labels)

    unique_label = np.unique(labels)

    L = []
    for i in range(len(unique_label)):
        L.append(i)
    L = np.array(L)
    label = []
    for x in os.listdir(path):
        tmp = labels[filenames == x[:-4]]

        tmp2 = L[unique_label == tmp]

        label.append(tmp2[0])

    label = np.array(label)

    return np.asarray(imgs, np.float32), np.asarray(label,np.int32)
```

```python
def minibatches(inputs=None, targets=None, batch_size=None, shuffle=False):
    assert len(inputs) == len(targets)
    if shuffle:
        indices = np.arange(len(inputs))
        np.random.shuffle(indices)
    for start_idx in range(0, len(inputs) - batch_size + 1, batch_size):
        if shuffle:
            excerpt = indices[start_idx:start_idx + batch_size]
        else:
            excerpt = slice(start_idx, start_idx + batch_size)
        yield inputs[excerpt], targets[excerpt]
```
